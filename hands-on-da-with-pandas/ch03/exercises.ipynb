```python
import pandas as pd 
appl_df = pd.read_csv('data/aapl.csv')
appl_df['ticker'] = 'AAPL'
appl_df.set_index('date', inplace=True)
appl_df.head()
```

```python
amzn_df = pd.read_csv('data/amzn.csv')
amzn_df['ticker'] = 'AMZN'
amzn_df.set_index('date', inplace=True)
amzn_df.head()
```

```python
fb_df = pd.read_csv('data/fb.csv')
fb_df['ticker'] = 'FB'
fb_df.set_index('date', inplace=True)
fb_df.head()
```

```python
goog_df = pd.read_csv('data/goog.csv')
goog_df['ticker'] = 'GOOG'
goog_df.set_index('date', inplace=True)
goog_df.head()
```

```python
nflx_df = pd.read_csv('data/nflx.csv')
nflx_df['ticker'] = 'NFLX'
nflx_df.set_index('date', inplace=True)
nflx_df.head()
```

```python
faang_df = pd.concat([appl_df, amzn_df, fb_df, goog_df, nflx_df])
faang_df.head()
```

```python
faang_df.to_csv('faang.csv')
```

```python
faang_df.index = pd.to_datetime(faang_df.index)
faang_df.sort_values(by=['date', 'ticker'], inplace=True)
faang_df.head(10)
```

```python
faang_df.nsmallest(7, 'volume').head(7)
```

```python
faang_df = faang_df.reset_index()
melted_faang = faang_df.melt(
    id_vars=['date', 'ticker'],
    value_vars=['open', 'high', 'low', 'close', 'volume'],
    value_name='value',
    var_name='attribute'
)

melted_faang.head()
```

```python
covid_df = pd.read_csv('data/covid19_cases.csv')
covid_df.head()
```

```python
covid_df['date'] = pd.to_datetime(covid_df['dateRep'], format='%d/%m/%Y')
covid_df.head()
```

```python
covid_df.set_index('date', inplace=True)
covid_df.sort_index(inplace=True)
covid_df.head()
```

```python
covid_df.replace('United_States_of_America', 'USA', inplace=True)
covid_df.replace('United_Kingdom', 'UK', inplace=True)
covid_df.loc[covid_df['countriesAndTerritories'] == 'United Kingdom']
covid_df.loc[covid_df['countriesAndTerritories'] == 'USA']
```

```python
covid_df.loc[(covid_df['countriesAndTerritories'] == 'USA') | (covid_df['countriesAndTerritories'] == 'UK')]
```

```python
p_df = covid_df.pivot(columns='countriesAndTerritories', values='cases').fillna(0)
p_df.head()
```
